```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lineplot(x="Gender",y="ProductID",data=a)
```

```python
a["PrimaryColor"].notnull()
```

```python
top_color=(a.groupby("PrimaryColor")["Price (INR)"].max().nlargest(10).index)
filtered=a[a["PrimaryColor"].isin(top_color)]           
sns.scatterplot(x="PrimaryColor",y="Price (INR)",data=filtered)
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
a.describe()
```

```python
sns.pairplot(data=a)
```

```python
top_Brand=(a.groupby("ProductBrand")["Price (INR)"].max().nlargest(10).index)
filtere=a[a["ProductBrand"].isin(top_Brand)]   
sns.boxplot(x="ProductBrand",y="Price (INR)",data=filtere)
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
a.fillna("Red")
```

```python
a_subset=a[["ProductName","Price (INR)"]]
top5=a_subset.nlargest(5,"Price (INR)")
sns.barplot(x=top5["ProductName"],y=top5["Price (INR)"])
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
g=a.drop(columns=["ProductBrand","Gender","Description","PrimaryColor"])

```

```python
g["sales"]=g["Price (INR)"]*g["NumImages"]
top10=g.nlargest(10,"sales")[["ProductName","sales"]]
sns.heatmap(top10.set_index("ProductName"),annot=True,fmt=".0f",cmap="OrRd")
plt.show()
```
